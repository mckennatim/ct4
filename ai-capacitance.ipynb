## How It Works (Brief Explanation)

A capacitor blocks DC because:
- DC voltage is constant, so once the capacitor charges to that voltage, no more current flows (like an open circuit)
- AC voltage constantly changes, so the capacitor continuously charges and discharges, allowing current to flow

## Calculating the Capacitor Value

The key formula is **capacitive reactance (Xc)**: 

```
Xc = 1 / (2πfC)
```

Where:
- Xc = capacitive reactance (Ω)
- f = frequency (Hz)
- C = capacitance (F)

**For AC Coupling Applications:**

The **cutoff frequency** (fc) determines which frequencies pass: 

```
fc = 1 / (2πRC)
```

Solving for C: 

```
C = 1 / (2πRfc)
```

Where:
- fc = cutoff frequency (-3dB point)
- R = load resistance (Ω)
- C = capacitance (F)

**Example:**
- To pass audio signals (20 Hz and above)
- With a 10kΩ load resistance
- Want cutoff at 20 Hz

```
C = 1 / (2π × 10,000 × 20)
C ≈ 0.8 µF
```

You'd typically use a 1 µF capacitor. 

**Rule of Thumb:** Choose fc to be 1/10th of your lowest frequency of interest for minimal signal loss.

This is a classic **AC coupling with DC bias** problem. You need to isolate the AC input from the DC bias circuit while allowing the AC signal to "ride on" the DC bias voltage.  Here's the solution:

## The Circuit

    VCC ----+------/\/\/\----+-----> To Next Stage
            |                |
            |                |
           ---              ---
    AC IN  ---  C1          --- C2 (optional)
    (0-1V)  |                |
    80Ω     |                |
            +----------------+
            |                |
           /\/\/\           /\/\/\
           R2 or Rbias      Rload
            |                |
           GND              GND

```

Better ASCII representation:

```

                           |
                           |

                        |
                        R2 (to GND)
```

## Detailed Circuit Diagram

```

 |          |

```

## How It Works

1. **R1 and R2** form a voltage divider that sets the DC bias point (independent of the AC source)
2. **C1** (coupling capacitor) allows AC to pass but blocks DC, preventing: 
   - The DC bias from affecting your AC source
   - Your AC source's low impedance (80Ω) from loading down the bias circuit
3. The AC signal passes through C1 and appears superimposed on the DC bias voltage

## Component Selection

### 1. DC Bias Resistors (R1, R2)

Choose values much larger than your 80Ω source impedance:
- Typical:  **10kΩ to 100kΩ**
- This prevents the AC source from loading the bias network
- The ratio R1:R2 sets your bias voltage

**Example for VCC = 5V, desired bias = 2.5V:**
- R1 = R2 = 47kΩ
- Bias voltage = VCC × R2/(R1+R2) = 5V × 0.5 = 2.5V

### 2. Coupling Capacitor (C1)

Calculate based on your lowest AC frequency: 

```
C1 = 1 / (2π × Rsource × fc)
```

Where:
- Rsource = 80Ω (your source impedance)
- fc = lowest frequency you want to pass

**Examples:**

| Frequency Range | fc | C1 Value |
|----------------|-----|----------|
| Audio (20 Hz+) | 20 Hz | **100 µF** |
| Voice (300 Hz+) | 300 Hz | **6.8 µF** |
| Radio/HF (1 kHz+) | 1 kHz | **2 µF** |
| High freq (10 kHz+) | 10 kHz | **0.22 µF** |

For audio with 80Ω source:
```
C1 = 1 / (2π × 80 × 20) ≈ 100 µF
```

Use **100 µF or larger** electrolytic (watch polarity!) or 100µF ceramic.

## Key Points

1. **High impedance bias network** (R1, R2 >> 80Ω) prevents AC source from loading it
2. **Large coupling cap** (C1) passes even low frequencies without attenuation
3. **DC isolation** - AC source sees only AC path, DC bias circuit sees only DC path
4. If C1 is electrolytic, positive terminal goes to the **higher DC voltage** (usually the bias point)

## Optional Improvements

1. **Add bypass capacitor** (0.1µF) from bias point to ground for AC stability
2. **Add series resistor** (small, like 100Ω) before C1 for protection
3. **Use non-polarized capacitor** if AC signal might reverse polarity

## Why I Initially Said Electrolytic

1. **High capacitance (470-680µF) in small package** - electrolytics are cheap and compact for large values
2. **Common in hobbyist circuits** - most tutorials show electrolytics for this application
3. **Cost** - electrolytics are much cheaper for large values

## Why You Might NOT Want Electrolytic

### 1. **Polarity Constraints**
- Must be installed correctly or they fail/explode
- If your CT could ever reverse (it shouldn't, but.. .), electrolytic won't like it

### 2. **Lifespan Issues**
- Electrolytics dry out over time (5-10 years typical)
- Temperature sensitive
- Not ideal for "install and forget" applications

### 3. **ESR (Equivalent Series Resistance)**
- Electrolytics have higher ESR than film caps
- Can affect signal quality slightly

### 4. **Leakage Current**
- Electrolytics leak more DC current
- Could slightly affect your bias point stability

## Better Alternatives

### **Film Capacitors (Polyester/MKT or Polypropylene)**

**Advantages:**
- **Non-polarized** - can be installed either way
- **Long life** - 20+ years
- **Low ESR** - better signal quality
- **Stable** - less temperature/aging drift
- **More reliable** for long-term installations

**Disadvantages:**
- **Larger physical size** - 470µF film cap is much bigger than electrolytic
- **More expensive** - typically 3-10x the cost

1-2µF polypropylene or polyester film capacitor

Your **load impedance** isn't just the 53Ω CT - it's the **parallel combination** of:
- CT:  53Ω

Since 60kΩ >> 53Ω, the AC impedance is essentially **53Ω**.

But actually, for the **cutoff frequency**, what matters is the impedance that C1 sees:

**C1 sees:** CT (53Ω) in series with [R1 || R2] = 53Ω + 60kΩ ≈ **60kΩ**

So the actual cutoff: 
```

```

Where R = R1 || R2 = 60kΩ (approximately, since 53Ω is negligible)

```
For fc = 5 Hz:
